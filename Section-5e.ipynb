## 5E. Trends by Time/ Day  


#### i. Average Active Minutes and Average Steps by Day of Week


<br>

The average active minutes were quite consistent and had small variation across the week, scoring between approximately 208 and 245 active minutes per day. Saturday had the most active minutes (244.27 mins) and Sunday had the lowest active minutes (208.49 mins). Tuesday and Saturday recorded the highest number of steps per day (> 8100 steps), while Sunday had the lowest recording (6933 steps). Looking at the graph, the disparity between active minutes and total steps was also consistent, indicating that as active minutes increase so do the total steps and vice versa. It's important to remember that the number of steps per day is also dependent on variables such as the type of exercise and users' average step length.  
<br>
#### ii. Sleep per Day of Week


<br>

In accordance with the U.S. Department of Health & Human Services, Saturday and Sunday met the recommended sleep. This is possibly due to people being less busy on the weekends. However, from Monday to Friday, users had insufficient sleep, averaging 410 minutes during workdays, which is on average 10 minutes less than the recommended.  
<br>
#### iii. Average Calories Burned by Day of Week


<br>

The heatmap shows that Tuesday and Saturday recorded the highest amount of calories burned, which is in accordance with the number of active minutes per day that also peaked on Tuesday and Saturday. Calorie expenditure is more widespread across Monday to Friday, as calorie expenditure starts to increase at 5 am. Comparatively, significant calorie expenditure occurred over fewer hours on the weekend. Calories burned on Saturday and Sunday started to increase at 7 am. This is a 2-hour difference compared to weekdays. The rate of calories burned per hour was greatest on Saturday at 1 pm and Wednesday at 5 pm. Calories burned on average peaked around 5-8 pm, possibly due to it being a popular time to exercise after work.  
<br>
#### iv. Average Heartrate by Time & Day of Week


<br>

In comparison to the calories burned heatmap, heart rate records were a bit more sporadic across the time and day of the week. Average heart rates had small variation across each day of the week. Heart rates on average were highest on Fridays at 79.01 BPM and were lowest on Sundays at 73.97 BPM. Additionally, there are slightly higher heart rate figures in the early mornings of Monday and Saturday. Between 4 pm and 9 pm, heart rates were consistently high, scoring above 81 BPM for each hour during this timeframe.  
<br>
#### vi. Average Intensity by Time and Day of Week


<br>

Average intensity had a wide amount of variation for each day of the week. Surprisingly, the average intensity was significantly higher on Tuesday compared to any other day of the week. Users' average Intensity was lowest on Sunday at 530. Similar to average calorie expenditure, intensity scores are more widespread across Monday to Friday, as Intensity starts to increase at 5 am and decreases rapidly at 11 pm. The timeframe with the highest intensities was from 5 pm to 8 pm, averaging over 320 per hour.
